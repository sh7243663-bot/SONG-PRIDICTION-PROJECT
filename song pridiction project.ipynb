```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestNeighbors
```

```python
import joblib
```

```python
df=pd.read_excel (r"C:\Users\hp\OneDrive\Documents\spotify_tracks_data.xlsx")
df['track_name']=df['track_name'].str.lower()
```

```python
print(df.head())
```

```python
feature_cols = [
    'energy',
    'danceability',
    'speechiness',
    'instrumentalness',
    'loudness',
    'liveness',
    'tempo',
    'valence',
    'acousticness']
```

```python
df= df.dropna(subset=feature_cols)
df
```

```python
scaler=StandardScaler()
scaled_features=scaler.fit_transform(df[feature_cols])

knn=NearestNeighbors(n_neighbors=6,metric='euclidean')
knn.fit(scaled_features)

joblib.dump(knn,"spotify_knn_model.pkl")
joblib.dump(scaler,"spotify_scaler.pkl")
```

```python
def get_song_features(song_name):
    song_name=song_name.lower()
    if song_name not in df['track_name'].values:
      return None
    song_row=df[df['track_name'] == song_name].iloc[0]
    return song_row[feature_cols].values.reshape(1,-1)
```

```python
input_songs = []
for i in range(5):
     song = input(f"Enter song {i+1} name:")
     features = get_song_features(song)
     
     if features is None:
        print("song is not foud dataset")
     else:
       input_songs.append(song)
```

```python
last_song = input_songs[-1] 
last_features = get_song_features (last_song)

last_features_scaled =scaler.transform(last_features)

distances, indices= knn.kneighbors (last_features_scaled)

recommended_songs =df.iloc[indices [0] [1:]]
```

```python
print("\n Recommended Songs Based on Your Last Input: \n")

for _, row in recommended_songs.iterrows():

    track = row['track_name']

    artist = row['artists'] if 'artists' in row else "Unknown Artist"

    print(f" {track} | {artist}")
```
